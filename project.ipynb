## Business understanding of a telecome indusrty cudtomers churn:

Understanding Customer Churn in the Telecom Industry

Customer churn poses a significant challenge across various industries, but it is particularly critical in the telecom sector. The impact on company revenue due to customer churn is substantial. As a result, telecom companies are actively working to develop predictive models for identifying potential churn.

In this fiercely competitive market, the telecommunications industry faces an average annual churn rate of 15-25%. Notably, acquiring a new customer costs significantly more—up to 5-10 times—than retaining an existing one. Consequently, customer retention has become a top priority, surpassing customer acquisition in importance.

To address this, it is crucial to identify the factors that contribute to customer churn. By understanding these drivers, companies can take necessary actions to reduce churn and enhance overall customer satisfaction.



the main goal of our project is to deveolpe an understanding of cause of the customer churn which assists telecome operators to predict customers who are most likely subject to churn, and what to do retain the most valuable customer.

#### Donut chart

```python
def plot_churn_distribution(data: list[int]):
    plt.pie(
        data,
        labels=['Not Churned', 'Churned'],
        colors=['cyan', 'fuchsia'],
        startangle=50,
        shadow=True,
        radius=2,
        explode=(0, 0.2),
        autopct='%1.2f%%',
        pctdistance=0.75
    )
    circle = plt.Circle((0, 0), 1, color='white')
    plt.gca().add_artist(circle)
    plt.title("Churned Vs Not churned Customers (DSL)", y = 1.4)
    plt.show()

plot_churn_distribution(A)
```

### CountPlot

```python
def plot_churn_count(data:pd.DataFrame, palette = 'husl'):
    sns.countplot(x='churn',data=data, palette = palette )
    plt.title('Churn count', y = 1)
    plt.xlabel('Churn')
    plt.ylabel('Count')
    plt.show()

plot_churn_count(data)
```

#### Count Plot

```python
def state_churn_count(data:pd.DataFrame, palette = 'husl', hue='churn'):
    data[hue] = data[hue].astype(str)
    plt.figure(figsize=(20, 7))
    sns.set(style='whitegrid')
    sns.countplot(x = 'state', data=data, palette=palette, hue=hue)
    plt.xticks(fontsize=13)
    plt.yticks(fontsize=13)
    
    plt.title('Churn/Not Churn State-Wise', fontsize=30, y = 1.1)
    
    plt.xlabel('States', fontsize=20, color='fuchsia')
    plt.ylabel('Count', fontsize=20, color='cyan')
    
    plt.show()

state_churn_count(data)
    
```

#### Line Chart

```python
def state_churn_rate(data:pd.DataFrame):
    x = data.state.unique()
    y = data.groupby('state')['churn'].mean()
    sns.set(style = 'darkgrid')
    plt.rcParams['figure.figsize'] = (20, 7)
    plt.rcParams['lines.linestyle'] = '-'
    plt.plot(x, y, color = 'black', marker = '*', markersize = '10')

    plt.title('State Churn Graph', fontsize=25)
    plt.xlabel('States', fontsize=20)
    plt.ylabel('Churn rate', fontsize=20);

state_churn_rate(data)
```

```python
dc_pm = data['total day charge'].mean()/data['total day minutes'].mean()
ec_pm = data['total eve charge'].mean()/data['total eve minutes'].mean()
nc_pm = data['total night charge'].mean()/data['total night minutes'].mean()
intcc_pm = data['total intl charge'].mean()/data['total intl minutes'].mean()
```

### Bar plot

```python
def diff_per_min_charge(data:pd.DataFrame, palette = 'husl'):
    sns.barplot(
        x=['day', 'evening', 'night', 'int'],
        y=[dc_pm, ec_pm, nc_pm, intcc_pm], 
        palette = palette
    )
    plt.title('Different per minute charge', y = 1 , fontsize=15)
    plt.xlabel('type', fontsize=15)
    plt.ylabel('Count', fontsize=15)
    plt.rcParams['figure.figsize']=(10,7)
    plt.xticks(fontsize=13)
    plt.yticks(fontsize=13)
    plt.show()
diff_per_min_charge(data)
        
    
```
